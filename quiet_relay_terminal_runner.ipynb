# Quiet Relay: 2026 - Data-Driven Terminal Battle Runner

This notebook runs the terminal combat core from `quiet_relay_terminal_datadriven.py`.

## How to use
1. Run the setup cell.
2. Edit the configuration cell.
3. Run the battle cell.
4. Optionally inspect the saved battle log.

## Notes
- This runner is for standalone battles, not the solo node-calibrated campaign flow.
- The campaign hardmode runner is `quiet_relay_vertical_slice_runner.ipynb`.
- Set `AUTO_MODE = True` for a full automated battle.
- Valid party members: `vanguard`, `duelist`, `cantor`, `ranger`, `penitent`.


````python
from pathlib import Path
import importlib.util
import sys

PROJECT_ROOT = Path.cwd()
SCRIPT_PATH = PROJECT_ROOT / 'quiet_relay_terminal_datadriven.py'
DEFAULT_LOG_PATH = PROJECT_ROOT / 'quiet_relay_last_battle_log_datadriven.txt'

if not SCRIPT_PATH.exists():
    raise FileNotFoundError(f'Could not find: {SCRIPT_PATH}')

print(f'Using script: {SCRIPT_PATH}')
print(f'Default log path: {DEFAULT_LOG_PATH}')


def load_quiet_relay_module():
    module_name = 'quiet_relay_terminal_datadriven_nb'
    if module_name in sys.modules:
        del sys.modules[module_name]
    spec = importlib.util.spec_from_file_location(module_name, SCRIPT_PATH)
    module = importlib.util.module_from_spec(spec)
    sys.modules[module_name] = module
    spec.loader.exec_module(module)
    return module


def run_quiet_relay(
    party='vanguard,duelist,cantor',
    scenario='bell_warden',
    auto_mode=True,
    seed=2026,
    log_file=DEFAULT_LOG_PATH,
):
    module = load_quiet_relay_module()
    argv = [
        '--party', party,
        '--scenario', scenario,
        '--seed', str(seed),
        '--log-file', str(log_file),
    ]
    if auto_mode:
        argv.append('--auto')

    print('Running with arguments:')
    print(' '.join(argv))
    print('-' * 72)
    rc = module.main(argv)
    print('-' * 72)
    print(f'Exit code: {rc}')
    return rc


def show_battle_log(log_file=DEFAULT_LOG_PATH):
    log_file = Path(log_file)
    if not log_file.exists():
        print(f'No log file found at: {log_file}')
        return
    print(log_file.read_text(encoding='utf-8'))
````

````python
# Edit these values, then run the battle cell below.

PARTY = 'vanguard,duelist,cantor'
SCENARIO = 'bell_warden'
AUTO_MODE = True
SEED = 2026
LOG_FILE = DEFAULT_LOG_PATH

print('Current configuration:')
print(' PARTY   =', PARTY)
print(' SCENARIO=', SCENARIO)
print(' AUTO    =', AUTO_MODE)
print(' SEED    =', SEED)
print(' LOG     =', LOG_FILE)
````

````python
# Run one battle
run_quiet_relay(
    party=PARTY,
    scenario=SCENARIO,
    auto_mode=AUTO_MODE,
    seed=SEED,
    log_file=LOG_FILE,
)
````

````python
# Inspect the saved log
show_battle_log(LOG_FILE)
````
