# Quiet Relay: 2026 - Data-Driven Solo Hardmode Runner

This notebook runs the data-driven campaign vertical slice from `quiet_relay_vertical_slice_datadriven.py`.

## What it includes
- Solo hardmode expedition flow
- Default solo character: `vanguard`
- Node-level Power / Precision / Composure calibration
- `--node-axis` fixed score support for autoplay smoke tests
- Optional `--axis-file` JSON support for per-node scores
- JSON save/load
- Per-battle logs and a run report

## Notes
- `AUTO_MODE = True` is easiest in Jupyter.
- `NODE_AXIS = (70, 70, 70)` gives a playable hardmode smoke run.
- Set `NODE_AXIS = None` and `AXIS_FILE` to a JSON path to use per-node calibration.
- `AUTO_MODE = False` will prompt once per node for Power, Precision, and Composure.


````python
import os
os.getcwd()
````

````python
from pathlib import Path
import importlib.util
import json
import sys

PROJECT_ROOT = Path.cwd()
SCRIPT_PATH = PROJECT_ROOT / 'quiet_relay_vertical_slice_datadriven.py'
SAVE_FILE = PROJECT_ROOT / 'quiet_relay_vertical_slice_datadriven_save.json'
LOG_DIR = PROJECT_ROOT / 'quiet_relay_vertical_slice_datadriven_logs'
RUN_REPORT = PROJECT_ROOT / 'quiet_relay_vertical_slice_datadriven_last_run.txt'

if not SCRIPT_PATH.exists():
    raise FileNotFoundError(f'Could not find: {SCRIPT_PATH}')

print(f'Using script: {SCRIPT_PATH}')
print(f'Save file:    {SAVE_FILE}')
print(f'Log dir:      {LOG_DIR}')
print(f'Run report:   {RUN_REPORT}')


def load_vertical_slice_module():
    module_name = 'quiet_relay_vertical_slice_datadriven_nb'
    if module_name in sys.modules:
        del sys.modules[module_name]
    spec = importlib.util.spec_from_file_location(module_name, SCRIPT_PATH)
    module = importlib.util.module_from_spec(spec)
    sys.modules[module_name] = module
    spec.loader.exec_module(module)
    return module


def run_vertical_slice(
    auto_mode=True,
    seed=2026,
    load_existing=False,
    fresh_start=False,
    solo_character='vanguard',
    node_axis=(70, 70, 70),
    axis_file=None,
    auto_route='first',
):
    module = load_vertical_slice_module()
    argv = [
        '--save-file', str(SAVE_FILE),
        '--log-dir', str(LOG_DIR),
        '--seed', str(seed),
        '--solo-character', str(solo_character),
        '--auto-route', str(auto_route),
    ]
    if auto_mode:
        argv.append('--auto')
    if load_existing:
        argv.append('--load')
    if fresh_start:
        argv.append('--fresh')
    if axis_file is not None:
        argv.extend(['--axis-file', str(axis_file)])
    elif node_axis is not None:
        power, precision, composure = node_axis
        argv.extend(['--node-axis', str(power), str(precision), str(composure)])
    print('Running with argv =', argv)
    return module.main(argv)


def write_axis_file(path, default=(60, 60, 60), nodes=None):
    path = Path(path)
    nodes = nodes or {}
    payload = {
        'default': {'power': default[0], 'precision': default[1], 'composure': default[2]},
        'nodes': nodes,
    }
    path.write_text(json.dumps(payload, indent=2), encoding='utf-8')
    print(f'Wrote axis file: {path}')
    return path


def show_save_file(path=SAVE_FILE):
    path = Path(path)
    if not path.exists():
        print('No save file present.')
        return
    payload = json.loads(path.read_text(encoding='utf-8'))
    print(json.dumps(payload, indent=2))


def show_latest_log(log_dir=LOG_DIR):
    log_dir = Path(log_dir)
    if not log_dir.exists():
        print('No log directory present.')
        return
    logs = sorted(log_dir.glob('*.txt'))
    if not logs:
        print('No battle logs yet.')
        return
    latest = logs[-1]
    print(f'Latest log: {latest}')
    print('-' * 80)
    print(latest.read_text(encoding='utf-8'))


def show_run_report(path=RUN_REPORT):
    path = Path(path)
    if not path.exists():
        print('No run report present.')
        return
    print(path.read_text(encoding='utf-8'))
````

````python
# Edit these values, then run the next cell.
AUTO_MODE = False
SEED = 2026
LOAD_EXISTING = False
FRESH_START = True
SOLO_CHARACTER = 'vanguard'
NODE_AXIS = None  # Set to None when using AXIS_FILE.
AXIS_FILE = None
AUTO_ROUTE = 'first'

print('Current configuration:')
print(' AUTO_MODE     =', AUTO_MODE)
print(' SEED          =', SEED)
print(' LOAD_EXISTING =', LOAD_EXISTING)
print(' FRESH_START   =', FRESH_START)
print(' SOLO_CHARACTER=', SOLO_CHARACTER)
print(' NODE_AXIS     =', NODE_AXIS)
print(' AXIS_FILE     =', AXIS_FILE)
print(' AUTO_ROUTE    =', AUTO_ROUTE)
print(' SAVE_FILE     =', SAVE_FILE)
print(' LOG_DIR       =', LOG_DIR)
print(' RUN_REPORT    =', RUN_REPORT)
````

````python
# Run the full data-driven vertical slice.
run_vertical_slice(
    auto_mode=AUTO_MODE,
    seed=SEED,
    load_existing=LOAD_EXISTING,
    fresh_start=FRESH_START,
    solo_character=SOLO_CHARACTER,
    node_axis=NODE_AXIS,
    axis_file=AXIS_FILE,
    auto_route=AUTO_ROUTE,
)
````
