# Quiet Relay: 2026 — Vertical Slice Runner

This notebook runs the campaign vertical slice from `quiet_relay_vertical_slice.py`.

## What it includes
- Hub prep and expedition flow
- One district: Rain Toll Corridor
- 3 playable characters: Vanguard, Duelist, Cantor
- 6 regular enemy types, 1 elite, 1 boss
- JSON save/load
- Per-battle logs and a run report

## Notes
- `AUTO_MODE = True` is easiest in Jupyter.
- `AUTO_MODE = False` will use terminal-style prompts inside the notebook.
- `LOAD_EXISTING = True` resumes from the JSON save file if it exists.
- `FRESH_START = True` forces a brand-new campaign.


````python
from pathlib import Path
import importlib.util
import json
import sys

SCRIPT_PATH = Path('/users/junhyeok/tbrg/quiet_relay_vertical_slice.py')
SAVE_FILE = Path('/users/junhyeok/tbrg/quiet_relay_vertical_slice_save.json')
LOG_DIR = Path('/users/junhyeok/tbrg/quiet_relay_vertical_slice_logs')
RUN_REPORT = Path('/users/junhyeok/tbrg/quiet_relay_vertical_slice_last_run.txt')

if not SCRIPT_PATH.exists():
    raise FileNotFoundError(f'Could not find: {SCRIPT_PATH}')

print(f'Using script: {SCRIPT_PATH}')
print(f'Save file:    {SAVE_FILE}')
print(f'Log dir:      {LOG_DIR}')
print(f'Run report:   {RUN_REPORT}')

def load_vertical_slice_module():
    module_name = 'quiet_relay_vertical_slice_nb'
    if module_name in sys.modules:
        del sys.modules[module_name]
    spec = importlib.util.spec_from_file_location(module_name, SCRIPT_PATH)
    module = importlib.util.module_from_spec(spec)
    sys.modules[module_name] = module
    spec.loader.exec_module(module)
    return module

def run_vertical_slice(auto_mode=True, seed=2026, load_existing=False, fresh_start=False):
    module = load_vertical_slice_module()
    argv = ['--save-file', str(SAVE_FILE), '--log-dir', str(LOG_DIR), '--seed', str(seed)]
    if auto_mode:
        argv.append('--auto')
    if load_existing:
        argv.append('--load')
    if fresh_start:
        argv.append('--fresh')
    print('Running with argv =', argv)
    return module.main(argv)

def show_save_file(path=SAVE_FILE):
    path = Path(path)
    if not path.exists():
        print('No save file present.')
        return
    with open(path, 'r', encoding='utf-8') as handle:
        payload = json.load(handle)
    print(json.dumps(payload, indent=2))

def show_latest_log(log_dir=LOG_DIR):
    log_dir = Path(log_dir)
    if not log_dir.exists():
        print('No log directory present.')
        return
    logs = sorted(log_dir.glob('*.txt'))
    if not logs:
        print('No battle logs yet.')
        return
    latest = logs[-1]
    print(f'Latest log: {latest}')
    print('-' * 80)
    print(latest.read_text(encoding='utf-8'))

def show_run_report(path=RUN_REPORT):
    path = Path(path)
    if not path.exists():
        print('No run report present.')
        return
    print(path.read_text(encoding='utf-8'))
````

````python
# Edit these values, then run the next cell.
AUTO_MODE = False
SEED = 2026
LOAD_EXISTING = True  # True for loading saved game
FRESH_START = False

print('Current configuration:')
print(' AUTO_MODE    =', AUTO_MODE)
print(' SEED         =', SEED)
print(' LOAD_EXISTING=', LOAD_EXISTING)
print(' FRESH_START  =', FRESH_START)
print(' SAVE_FILE    =', SAVE_FILE)
print(' LOG_DIR      =', LOG_DIR)
print(' RUN_REPORT   =', RUN_REPORT)
````

````python
# Run the full vertical slice
run_vertical_slice(auto_mode=AUTO_MODE, seed=SEED, load_existing=LOAD_EXISTING, fresh_start=FRESH_START)
````
